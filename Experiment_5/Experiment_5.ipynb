Problem statement 3: One set contains students in the Cricket Club another in football club
find students who are: 1. In both clubs 2. In only one club

Concept Used: Intersection, Symmetric difference

```python
cricket = {"amey", "raj", "raju", "aman", "jai", "dhvan","virat" }
football ={"aditya", "virat", "rohit", "dhvan", "kapil"}
#Intersection
print("Students in both clubs:", cricket & football)
#Symmetric Difference
print("Students in only one club:", cricket ^ football)
```

Problem Statement 4: Absent students
Given a set all students and a set of present students, find the absent students.

Concept used: Set difference (-)

```python
all_students = {"A", "B", "C", "D", "E"}
present_students = {"A", "C", "E"}

absent_students = all_students - present_students

print("All students:", all_students)
print("Present students:", present_students)
print("Absent students:", absent_students)
```

Problem Statement 5: Remove invalid entries. A set contains valid course code. Remove a discontinued course code from the set

Concept used: remove()/discard()

```python
valid_courses = {"CS101", "MA201", "PH301", "BI401", "CH501"}
discontinued_course = "PH301"

print("Original valid courses:", valid_courses)

valid_courses.remove(discontinued_course)

print("Valid courses after removing", discontinued_course + ":", valid_courses)

valid_courses.discard("AR101")
print("Valid courses after trying to discard non-existent AR101:", valid_courses)

valid_courses.discard("CS101")
print("Valid courses after discarding CS101:", valid_courses)
```

2. A dictionary stores student names and marks
- Write a python program to search for a student's marks using the name entry
- If the student is not found, display a suitable message.

```python
student_marks = {
    "A": 85,
    "B": 92,
    "C": 78,
    "D": 95,
    "E": 88
}

search_name = input("Enter the student's name to search for marks: ")

if search_name in student_marks:
    print(f"{search_name}'s marks: {student_marks[search_name]}")
else:
    print(f"Student '{search_name}' not found in the record.")
```

3. A dictionary stores usernames and passwords.
- Write a python program to validate user login

```python
usernames_passwords = {
    "user1": "password1",
    "user2": "password2",
    "user3": "password3"
}

username_input = input("Enter your username: ")
password_input = input("Enter your password: ")

stored_password = usernames_passwords.get(username_input)

if stored_password == password_input:
    print("Login successful!")
else:
    print("Invalid username or password.")

```

4. A dictionary stores student names and marks
- Write a Python program to find the student who scored the highest marks

```python
marks = {"A": 85, "B": 92, "C": 78, "D": 95, "E": 88}
topper = max(marks, key=marks.get)
print("Topper:", topper)
print("Marks:", marks[topper])
```
