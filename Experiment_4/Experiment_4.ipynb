```python
#Python tuples
#Access tuple items
thistuple = ("apple", "banana", "cherry")
print(thistuple[1])
print(thistuple)
```

```python
thistuple = ("apple", "banana", "cherry")
thistuple[1] = "strawberry"
print(thistuple)
#'tuple' object does not support item assignment
```

```python
#Add tuple to a tuple
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y
print(thistuple)
```

```python
##Adding tuple to tuple and in this y is stored as a address and when added it gets added to the address where += y shows that
#Address Verification after appending new items in the tuple
thistuple = ("apple", "banana", "cherry")
print("Before:",id(thistuple))
y = ("orange",)
thistuple += y
print("After:",id(thistuple))
```

Write a python program to:
Store the attendance record in the table
Count and display the total number of present days
Count and display the total number of absent days
Check whether the employee was absent atleast once during the recorded period and display an appropiate message.

"P" represents Present "A" represents Absent

```python
attendance = ("P","A","P","A","P","P","A")
print(attendance)
print()
present_count = attendance.count("P")
print(f"total number of present days: {present_count}")
print()
absent_count = attendance.count("A")
print(f"total number of absent days: {absent_count}")
print()

if attendance.count("P") >=1:
    print("was absent atleast once")
else:
    print("was not absent atleast once")
print()





```
