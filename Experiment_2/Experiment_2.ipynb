```python
# taking 5 subjects as float to allow decimals
s1 = float(input("subject 1: "))
s2 = float(input("subject 2: "))
s3 = float(input("subject 3: "))
s4 = float(input("subject 4: "))
s5 = float(input("subject 5: "))

total = s1 + s2 + s3 + s4 + s5
average = total / 5

print(f"sum: {total}, average: {average}")
```

```python
marks = [float(input(f"subject {i+1}: ")) for i in range(5)]
print(f"sum: {sum(marks)}, average: {sum(marks)/len(marks)}")
```

```python
# area of circle
import math
radius = float(input("enter radius: "))
area = math.pi * pow(radius, 2)
print(f"circle area: {area:.2f}")
```

```python
# area and perimeter of rectangle
length = float(input("enter length: "))
width = float(input("enter width: "))
area = length * width
perimeter = 2*(length+width)
print(f"rectangle area: {area: .2f}")
print(f"perimeter:{perimeter: .2f}")


```

```python
#simple calculator
num1 = float(input("enter first number: "))
num2 = float(input("enter second number: "))
op = input("enter operator(+,-,*,/): ")
if op == "+":result = num1 + num2
elif op == "-":result = num1-num2
elif op == "*":result = num1*num2
elif op == "/":result = num1/num2 if num2 != 0 else "error"
print(f"{num1} {op} {num2} = {result}")

```
